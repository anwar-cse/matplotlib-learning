Basic line chart

```python
import matplotlib.pyplot as plt

x=[1,2,3,5]
y=[4,5,6,7]

plt.plot(x,y)
plt.show()
```

Title and level

```python

plt.plot(x,y)
plt.title("Simple line chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

Style change

```python
plt.plot(x,y,linestyle="--",marker="*")
plt.show()
```

legend

```python
y2=[7,8,9,15]
plt.plot(x,y,label="data1")
plt.plot(x,y2,label="data2")
```
