Basic pie chart

```python
import matplotlib.pyplot as plt

label=["a","b","c"]
size=[20,50,30]
plt.pie(size,labels=label)
plt.show()
```

percentage show

```python
plt.pie(size,labels=label,autopct="%1.1f%%")
plt.show()
```

highlight

```python
explode=[0,0.5,0]
plt.pie(size,labels=label,autopct="%1.1f%%",explode=explode)
plt.show()
```

subplot

```python
import matplotlib.pyplot as plt

# 1st graph
plt.subplot(1,2,1)
plt.plot([1,2,3],[10,20,30])
plt.title("Line")

# 2nd graph
plt.subplot(1,2,2)
plt.bar(["A","B"], [5,7])
plt.title("Bar")

plt.show()
```
