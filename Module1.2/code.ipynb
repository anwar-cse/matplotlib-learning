Basic Bar chart

```python
import matplotlib.pyplot as plt

name=["anwar","rakib","habib"]
marks=[3.76,3.61,3.56]

plt.bar(name,marks)
plt.show()
```

title and level and width

```python
plt.bar(name,marks,width=0.3)
plt.title("marks of students")
plt.xlabel("name")
plt.ylabel("marks")
```

Horizontal bar chart

```python
plt.barh(name,marks)
plt.show()
```

Histogram

```python
import numpy as np
data=np.array([10,20,10,40,50])

plt.hist(data)
plt.show()
```
